# Making blobs
We will make blobs using the Sklearn make blobs method. This allows us to create random clusters of data. We will create 50 points, each with an x and y coordinate (two features), and will have four centers. We can also vary the standard deviation of our clusteres. Below we have a standard deviation of .5, so our clusters are tightly packed.

```python
X, y = make_blobs(n_samples=50, n_features=2, centers=4, cluster_std=.5, random_state=16)

def colors(y):
    if y == 0:
        return "red"
    elif y == 1:
        return "lightseagreen"
    elif y == 2:
        return "magenta"
    else:
        return "blue"

c = [colors(label) for label in y]

plt.scatter(X[:, 0], X[:, 1], color = c, alpha=0.6)
plt.show()
```

With a standard deviation of 3, we have a lot of variation in our clusters leading to a lot of overlap.

```python
X, y = make_blobs(n_samples=50, n_features=2, centers=4, cluster_std=3, random_state=16)

def colors(y):
    if y == 0:
        return "red"
    elif y == 1:
        return "lightseagreen"
    elif y == 2:
        return "magenta"
    else:
        return "blue"

c = [colors(label) for label in y]

plt.scatter(X[:, 0], X[:, 1], color = c, alpha=0.6)
plt.show()
```

The standard deviation of our centers will be $\sqrt{2}$ giving us a clean variance of 2. Here our clusters are still distinguishable but also are able to get near each other -- reflecting real world noise.

```python
X, y = make_blobs(n_samples=50, n_features=2, centers=4, cluster_std=np.sqrt(2), random_state=16)

def colors(y):
    if y == 0:
        return "red"
    elif y == 1:
        return "lightseagreen"
    elif y == 2:
        return "magenta"
    else:
        return "blue"

c = [colors(label) for label in y]

plt.scatter(X[:, 0], X[:, 1], color = c, alpha=0.6)
plt.show()
```

# K-Means Clustering
To implement the algorithm, we need to functions that are capable of doing the following:
* Pick K centroids from our dataset (initialize centroids)
* Assign each point to the nearest centroid and give it that label (assign clusters)
* For each group, assign a new centroid to average (update centroids)

We will then write a class that combines these methods and allows us to visualize our clustering.

```python
def initialize_centroids(X, k):
    """
    Creates initial random centroid from dataset (X) and k centroids
    """
    # Get k random indicies from data and return there corresponding points
    indices = np.random.choice(X.shape[0], k, replace=False) #avoid duplicate points
    
    return X[indices]

def assign_clusters(X, centroids):
    """
    Assign clusters to data based on clusters
    """
    cluster_assignments = []
    
    for point in X:
        # Take distance from each point to eachcentroid
        distances = [np.linalg.norm(point - centroid) for centroid in centroids]
        
        # Assign that point to its closest centroid
        closest_cluster = np.argmin(distances)
        cluster_assignments.append(closest_cluster)
    
    return np.array(cluster_assignments)



def update_centroids(X, labels, k):
    """
    Creates new centroids based on data X, labels of clustering, and number of 
    clusters k
    """
    return np.array([X[labels == i].mean(axis=0) for i in range(k)])

```

```python
class KMeans(object):

    def __init__(self, seed = None):
        np.random.seed(seed)
    
    def train(self, X, k, max_iters = 100):
        self.X = X
        
        centroids = initialize_centroids(X, k)

        for count in range(max_iters):
            labels = assign_clusters(X, centroids)
            new_centroids = update_centroids(X, labels, k)

            if np.all(centroids == new_centroids):
                break  # Stop once centroids converge

            centroids = new_centroids

        self.centroids = centroids
        self.labels = labels
        return labels, centroids

    def visualize(self):
        plt.scatter(self.X[:, 0], self.X[:, 1], c=self.labels, cmap='viridis')
        plt.scatter(self.centroids[:, 0], self.centroids[:, 1], s=10, c='red', marker='X', label='Centroids')
        plt.legend()
        plt.title("K-Means Clustering")
        plt.show()    
```

```python
model = KMeans(16)
labels, centroids = model.train(X,4)
model.visualize()
```

### A Few Potential Metrics
Two ideas I had are the following
* Maximal internal cluster distance -- as the name suggests, this is the largest distance between two points inside any cluster. This is very context dependent but the underlying idea is that you do not want points that are too different in the same cluster.
* Minimal centroid distance -- as the name suggests, this is the smallest distance between two centroids. This is also context dependent, but aims to conteract our other metric by ensuring that the centroids do not get too close and thus become redundant.


```python
def max_internal_cluster_distance(X, labels, centroids):
    max_dist = 0
    for label in set(labels):
        cluster_dist = np.max(np.linalg.norm(X[labels == label] - centroids[label], axis=1))
        max_dist = max(cluster_dist, max_dist)
    return max_dist
```

```python
def minimal_centroid_distance(centroids):
    num_centroids = centroids.shape[0]
    min_dist = np.inf

    for i in range(num_centroids):
        for j in range(i + 1, num_centroids): 
            distance = np.linalg.norm(centroids[i] - centroids[j])
            min_dist = min(min_dist, distance)

    return min_dist
```

```python
max_dists = []
min_dists = []
for k in range(1, 25):
    model = KMeans()
    labels, centroids = model.train(X,k)
    max_dists.append(max_internal_cluster_distance(X, labels, centroids))
    min_dists.append(minimal_centroid_distance(centroids))


# Create the plot
plt.plot(min_dists, label='Min Distances', marker='o')
plt.plot(max_dists, label='Max Distances', marker='s')

# Add labels and title
plt.xlabel('k')
plt.ylabel('Value')
plt.title('Min vs Max Distances')

# Show legend
plt.legend()

# Display the plot
plt.show()
```

So as we can see that the two lines of the metrics intersect for the first time at $k=4$. When the values of the two metrics get very close (or intersect), it means that worst case distances inside and between centroids are the same. Using the smallest value of $k$ when this is the case is probably a good idea because it creates the simplest model when this is the case.

### Another Potential Metrics
Instead of just considering worst case scenarios (as with the two above metrics), I also propose the following metric to describe a more average case.
* External Variability - the variability in distancies between centroids. We want a mid sized value (the scale is determined by calculating it for many values of k). If it is too small, then we have too few clusters. If it is too large, then we have too many clusters that vary a lot in distance from one another.

```python
def external_variability(centroids):
    num_centroids = centroids.shape[0]
    variabilities = []
    
    for i in range(num_centroids):
        for j in range(i + 1, num_centroids): 
            variabilities.append(np.linalg.norm(centroids[i] - centroids[j]))

    return np.var(variabilities)
```

```python
ext_var = []
for k in range(1, 25):
    model = KMeans()
    labels, centroids = model.train(X,k)
    ext_var.append(external_variability(centroids))


# Create the plot
plt.plot(ext_var, label='External Var', marker='s')

# Add labels and title
plt.xlabel('k')
plt.ylabel('Value')
plt.title('Ext Variability Measure vs k')

# Show legend
plt.legend()

# Display the plot
plt.show()
```

We want a medium sized external variability measure. So it looks like it maxes out at around 20. So half of that is 10 and that gives us $k=3,4,5$ as the closest values. We could also choose $k=4$ because it has a minimum value before a large increase in variability. This means that it is the largest value of $k$ before a lot the centers become spaced too inconsistently.

### Testing External Variability on Some Other Clusters

```python
def test_ext_var(clusters, std):
    X, y = make_blobs(n_samples=50, n_features=2, centers=clusters, cluster_std=std, random_state=16)
    
    plt.title("Clusters")
    plt.scatter(X[:, 0], X[:, 1], c = y, cmap="viridis", alpha=0.6)
    plt.show()
    
    ext_var = []
    for k in range(1, 31):
        model = KMeans()
        labels, centroids = model.train(X,k)
        ext_var.append(external_variability(centroids))

    # Create the plot
    plt.plot(ext_var, label='External Var', marker='s')

    # Add labels and title
    plt.xlabel('k')
    plt.ylabel('Value')
    plt.title(f'Ext Variability with {clusters} clusters and std of {std}')

    # Show legend
    plt.legend()

    # Display the plot
    plt.show()

test_ext_var(8,2)
```

```python
test_ext_var(5,5)
```

I really like this method for determining $k$ with external variability. With 8 clusters and variability of 2, we get a max score of 50. A medium score would be half of that (25) which corresponds $k=8,9$ with 9 as the local minumum.

With fewer clusters and higher variability, we can a max score of 60. Half of that is 30 which corresponds to $k=4,5$. But, at $k=4$, the graph is at a minimum value. If we choose $k=5$, then it owuld also be closer to our medium value of 30.

The only potential draw back with this method is that you have to test it on a large range of $k$ values and also compare a few different values of $k$ that are close to the target range. The score could also vary depeding on the seeding of K-Means.

# Takeaways
Creating metrics for k-means clustering is hard. While the three methods I proposed represent distinct elements of clustering, this is a complicated multifaceted problem which will need to rely on human intuiton.

For example, consider the below scenario where the actual data is the following three clusters. One large one with higher internal variability and two smaller ones that have small internal variability but are relatively close together.

![Challenging Scenario](Challenging_Scenario.png)

K-Means Clustering would have a hard time classifying this but our metrics would have an even harder time quantifying the performance. These algorithms and scores would seek to split the large blob into multiple clusters and keep the bottom two clusters as one cluster until many $k$ values are used. As such, outside knowledge of the problem and visual confirmation are useful in verifying the results. The metrics are a useful starting point, but ultimately this is a multifaceted problem that requires a lot of human insight.
