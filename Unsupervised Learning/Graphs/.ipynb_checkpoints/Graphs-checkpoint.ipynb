```python
import networkx as nx
import matplotlib.pyplot as plt
import random
from collections import Counter, defaultdict
import pandas as pd
import numpy as np
from ortools.linear_solver import pywraplp
```

```python
df = pd.read_csv("CFB_Schedule_2024.csv")

# Only want regular season
df = df[df.Wk <= 16]

#Get rid of top 25 ranking designators
df.Winner = df.Winner.str.replace(r'\(\d+\)\s*', '', regex=True)
df.Loser = df.Loser.str.replace(r'\(\d+\)\s*', '', regex=True)
```

Only want regular season
Only want teams that played more than 3 games (elimates all FCS teams that would over populate data and complicate results)

```python
# Combine counts for both columns
# Only want teams that played more than 3 games
combined_counts = pd.concat([df["Winner"], df["Loser"]]).value_counts(ascending = True)
smaller_teams = combined_counts[combined_counts < 3].index


# Drop rows where either Winner or Loser is in smaller_teams
df = df[~df["Winner"].isin(smaller_teams) & ~df["Loser"].isin(smaller_teams)]
```

# Creating our Graph

```python
G = nx.Graph()

for _, row in df.iterrows():
    G.add_edge(row['Winner'], row['Loser'])


plt.figure(figsize=(6, 6))
nx.draw(G, with_labels=True, node_color="lightblue", edge_color="gray", node_size=3000, font_size=10)
plt.show()
```

```python
print(len(G.nodes))
```

Because we have 134 nodes (which is the number of FBS teams), the visualization is very cluttered. We will create a visualization that focuses on the subset of teams including Rice and the teams they played.

```python

target_node = "Rice"

# Get neighbors
neighbors = list(G.neighbors(target_node))

# Create a subgraph containing the target node and its neighbors
sub_G = G.subgraph([target_node] + neighbors)
pos = nx.kamada_kawai_layout(sub_G)  # Alternative layout

plt.figure(figsize=(10, 10))  # Increase sizepos = nx.spring_layout(sub_G)  # Layout for better spacing
nx.draw(sub_G, pos, with_labels=True, node_color="lightblue", node_size=2500, font_size=10)
nx.draw_networkx_nodes(sub_G, pos, nodelist=[target_node], node_color="red", node_size=3000)
plt.show()
```

# Label Propogation
The fundamental principal of Label Propogation is that nodes tend to adopt the community (label) that the majority of their neighbors belong to. So given an undirected graph $G=(V,E)$:
1. __Initialize__: Assign each node $v\in V$  a unique label $L(v)=v$
2. __Iterate__: At each iteration, every node $v\in V$ updates its label $L(v)\leftarrow argmax|\{u\in N(v): L(u)=l\}|$
    * Ties are broken randomly
    * Repeat until convergence or max number of iteratiions is reached

```python
def flip_dict(dictionary):
    """
    Flips a dictionary so that the values become the keys
    and the keys become the values as a list
    """
    flipped_dict = defaultdict(list)
    for key, value in dictionary.items():
        flipped_dict[value].append(key)

    return dict(flipped_dict)
```

```python
class LabelPropagation():
    def __init__(self, max_iter = 100, seed = None):
        self.max_iter = max_iter
        self.seed = seed
        
    def label_propagation(self, G):
        # Initialize labels
        
        self.stages = []
        random.seed(self.seed)

        labels = {node: node for node in G.nodes()}

        for iteration in range(self.max_iter):
            changed = False
            nodes = list(G.nodes())
            random.shuffle(nodes)  # Random order for fairness
            for node in nodes:
                # Count neighbors' labels
                neighbor_labels = [labels[neighbor] for neighbor in G.neighbors(node)]
                if neighbor_labels:
                    # Most common label among neighbors
                    most_common_label, _ = Counter(neighbor_labels).most_common(1)[0]
                    if labels[node] != most_common_label:
                        labels[node] = most_common_label
                        changed = True
            
            self.stages.append(labels.copy())
            if not changed:
                print(f"Converged after {iteration+1} iterations.")
                break
        return labels
    
    def show_convergence(self):
        
        lens = []
        its = []
        for i, stage in enumerate(self.stages):
            community_sizes = []
            for value in flip_dict(stage).values():
                community_sizes.append(len(value))
            its.append(i)
            lens.append(community_sizes)
        
        df = pd.DataFrame(lens)
        
        fig, ax = plt.subplots(figsize=(8, 6))
        # Define x labels (each index)
        x_labels = range(len(df))
        bottom = np.zeros(len(df))

        for i in range(df.shape[1]):  # Iterate over columns
            ax.bar(x_labels, df.iloc[:, i], bottom=bottom, label=f'Component {i}')
            bottom += df.iloc[:, i]  # Update stacking position

        unique_stacks = list(df.count(axis=1))

        # Annotate bars with the count of unique stacks
        for i, count in enumerate(unique_stacks):
            ax.text(i, bottom[0] + 1, str(count), ha='center', fontsize=12, fontweight='bold')


        plt.title("Community Size Per Iteration", pad = 20)
        plt.xlabel("# Iterations")
        plt.ylabel("Community Size")
        plt.show()
        
    def pretty_print(self):

        flipped_dict = flip_dict(self.stages[-1])

        for i, community in enumerate(flipped_dict):
            print(f"Community {i+1}: {flipped_dict[community]}")
            
model = LabelPropagation(seed = 10)
model.label_propagation(G)
model.pretty_print()
model.show_convergence()
```

### Independence Number
What is the largest subset of nodes in our graph that are disconnected (an independent set of nodes). 

Let $x_i\in \{0,1\}$ denote if node $i$ is in our independent set. We want to solve the following model:


$$
\begin{align*} 
 \max &\sum_{i \in V} x_i \\ 
 \text{s.t.}\quad &x_i + x_j \leq 1, \quad \forall (i,j) \in E \\ 
 &x_i \in \{0,1\}, \quad \forall i \in V 
\end{align*}
$$

```python
def max_indepdendent_set(V, E):
    solver = pywraplp.Solver.CreateSolver('SCIP')

    # Define binary decision variables x_i
    x = {i: solver.BoolVar(f'x_{i}') for i in V}

    # Objective: Maximize sum of x_i
    solver.Maximize(solver.Sum(x[i] for i in V))

    # Constraints: x_i + x_j <= 1 for all edges (i, j) in E
    for i, j in E:
        solver.Add(x[i] + x[j] <= 1)

    # Solve the problem
    status = solver.Solve()

    if status == pywraplp.Solver.OPTIMAL:
        independent_set = [i for i in V if x[i].solution_value() == 1]
        return independent_set
    else:
        return None

solution = max_indepdendent_set(G.nodes, G.edges)
print("Maximum Independent Size:", len(solution))
for team in solution:
    print(team)
```

This is the largest group of schools all of which did not play each other in the 2024 season.

### Max Clique Size

Let $x_i\in \{0,1\}$ denote if node $i$ is in our clique. We want to solve the following model:


$$
\begin{align*} 
 \max &\sum_{i \in V} x_i \\ 
 \text{s.t.}\quad &x_i + x_j \leq 1, \quad \forall (i,j) \notin E \\
 &x_i \in \{0,1\}, \quad \forall i \in V 
\end{align*}
$$



```python
def max_clique_set(V, E):
    solver = pywraplp.Solver.CreateSolver('SCIP')

    # Define binary decision variables x_i
    x = {i: solver.BoolVar(f'x_{i}') for i in V}

    # Objective: Maximize sum of x_i
    solver.Maximize(solver.Sum(x[i] for i in V))

    # Constraints: x_i + x_j <= 1 for all non-adjacent pairs (i, j)
    non_edges = {(i, j) for i in V for j in V if i != j and (i, j) not in E and (j, i) not in E}
    for i, j in non_edges:
        solver.Add(x[i] + x[j] <= 1)

    # Solve the problem
    status = solver.Solve()

    if status == pywraplp.Solver.OPTIMAL:
        clique_set = [i for i in V if x[i].solution_value() == 1]
        return clique_set
    else:
        return None

solution = max_clique_set(G.nodes, G.edges)
print("Maximum Clique Size:", len(solution))
for team in solution:
    print(team)
```

This is the largest group of schools which all played each other in the 2024 season.
