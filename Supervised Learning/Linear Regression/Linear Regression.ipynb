# What offensive metrics correlate the most to points?

Our goal here is to predict how many points a player scored based off of the following metrics:
* Minutes
* Field Goals Attempted (non 3 pointers)
* Three Points Attempted
* Free Throws Attempted
* Assists    

To estimate how these variables impact the number of points a player scores in a game, we will assume there is a linear relationship between these factors and points scored.

---
# Linear Regression
The heart of linear regression is to predict our $i$-th depedent value $y^i$ as $\hat{y}^i$ with the following formula 
$$\hat{y}^i=w^Tx+b$$
for our array of features $x$. In order to do this, we want to minimize the mean square error between our predictions and our observed values.

$$C(w,b)=\frac{1}{2N}\sum_{i=1}^N(\hat{y}^i-y^i)^2$$
In order to do this, we will use gradient descent
## Establishing our gradient
Our cost can be rewritten by plugging in our regression into it
$$C(w,b)=\frac{1}{2N}\sum_{i=1}^N(w^Tx+b-y^i)^2$$
So we get that our gradient is
$$\nabla C(w,b)=(\frac{\partial C}{\partial w},\frac{\partial C}{\partial b})=((\hat{y}^i-y^i)x,\hat{y}^i-y^i)$$
## Descent
Now what we want to do is head down our gradient to the minimum of $C(w,b)$. We can do this by going through each sample in our training set and calculating the error for that sample. Then you adjust your weights according to the above some predetermined learning rate $\alpha$ and the above gradient formulas.  
For all $i$ corresponding to our samples
$$w \leftarrow w - \alpha (\hat{y}^i-y^i)x$$
$$b \leftarrow b - \alpha (\hat{y}^i-y^i)$$
This will give us our linear regression.

# Data Cleaning
We will start by reading in our data and removing the FGA column. This is because field goal attempts includes two and three point attempts and this double counts with FG3A (three point attempts). Instead, we will introduce a new column FG2A representing the number of non-three point (ie two point) field goal attempts.

# Training and Implementing the Model
Notice how the underlying code for the linear regression is a SingleNeuron similar to the perceptron. It just varies in its linear activation function

```python
X = df[X_factors].values
y = df.PTS.values

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=16)
```

```python
class SingleNeuron(object):
    def __init__(self, activation_function):
        """
        Initializes the activation function
        """
        self.activation_function = activation_function

    def train(self, X, y, alpha = 0.005, epochs = 50):
        """
        Trains the linear regression on X features and y results.
        Also uses a learning rate alpha and number of epochs.
        """
        self.w_ = np.random.randn(1 + X.shape[1])
        self.errors_ = []
        N = X.shape[0]

        for _ in range(epochs):
            errors = 0
            for xi, target in zip(X, y):
                error = (self.predict(xi) - target)
                self.w_[:-1] -= alpha*error*xi
                self.w_[-1] -= alpha*error
                errors += .5*(error**2)
            self.errors_.append(errors/N)
        return self

    def predict(self, X):
        """
        Predicitons on array X
        """
        preactivation = np.dot(X, self.w_[:-1]) + self.w_[-1]
        return self.activation_function(preactivation)

def linear_activation(z):
    return z
```

```python
node = SingleNeuron(linear_activation)
node.train(X_train, y_train, alpha = 0.0001, epochs = 10)
```

```python
plt.figure(figsize = (10, 8))
plt.plot(range(1, len(node.errors_) + 1), 
         node.errors_,
         marker = "o",
         label = "MSE")
plt.xlabel("epochs", fontsize = 15)
plt.ylabel("MSE", fontsize = 15)
plt.xticks(range(1, len(node.errors_) + 1))
plt.legend(fontsize = 15)
plt.title("MSE Error at Each Epoch During Training", fontsize = 18)
plt.show()
```

```python
mse = 0
h = []
for X_test_case, y_true in zip(X_test, y_test):
    mse += (y_true - node.predict(X_test_case)) ** 2
    h.append((y_true - node.predict(X_test_case)))
mse /= len(X_test)
plt.hist(h)
plt.title("Actual Points minus Predicted Points")
plt.show()
```

```python
print(f"Root Mean Squared Error of predictions: {mse ** .5}")
print(f"Standard Deviation of Test Values (Sample): {np.std(y_test, ddof=1)}")
print(f"Standard Deviation of all points (Population): {np.std(y)}")
```

# Analysis

Our root mean square error on our testing data is 3.5 points. This is significantly less than the sample standard deviation of the testing data and the population standard deviation of all data both of which are around 9. This is a good sign that our model is predicting the amount of points a player will score better than the random variability of the data. Likewise observing our above histogram of our errors, we can see that most of our data is between plus or minus 5 error, again within the standard deviation of the data. It is also worth noting the normal distribution of our errors.
