# Centers from Season Metrics
Logistic regression is like our perceptron but has a more robust probabilistic underpinnings. Mainly, it can determine the probability that a data point is in one of two categories, rather than classifying it based on some linear division.

As such, we will revist our problem from the perceptron but from a more complex angle. We will try to identify if a player is a center or not based on their season. Specefically, we are going to be considering the player's average (of games when they play over a quarter) for
* Three Point Attempts
* Rebounds
* Blocks
* Field Goal Percentage

---
# Logistic Regression
Logistic Regression has the same underlying principles as the perceptron and linear regression. What makes it unique is its activation function. The sigmoid activation function takes a real number and returns an output in the range $(0,1)$ -- which can represent the probability of being in the target category we are trying to classify.
$$
\sigma(z) = \frac{1}{1 + e^{-z}}
$$



So the probability of being a center given a set of stats for player $(i)$ is going to be
$$
P\Big(y^{(i)}\mid x^{(i)}\Big)=\begin{cases}
          \hat{y}^{(i)}, \quad &  y^{(i)} = 1 \\
          1-\hat{y}^{(i)}, \quad & y^{(i)} = 0 \\
     \end{cases}
$$
where $\hat{y}^{(i)}$ is our prediction and 1 represents a center.

---

Our goal is to maximize the sum of all our probabalities. Likewise, we can minimize the negative of the sum and likewise the negative average, which we define below with an equivilent expression called the binary cross entropy:

$$
L(\mathbf{w}, b) = -\frac{1}{N} \sum_{i=1}^{N} P\Big(y^{(i)}\mid x^{(i)}\Big) = \frac{1}{N}\sum_{i=1}^{N}\Big[ -y^{(i)}\log \hat{y}^{(i)} - (1-y^{(i)}) \log (1 - \hat{y}^{(i)})\Big ]
$$
Using some calculus, we can find the gradient of the binary cross entropy function
For all $i$ corresponding to our samples
$$\frac{\partial C(w_1, b; x^{(i)},y^{(i)})}{\partial w_1} = (\hat{y}^{(i)} - y^{(i)}) x^{(i)}$$
and

$$ \frac{\partial C(w_1, b; x^{(i)},y^{(i)})}{\partial b} = (\hat{y}^{(i)} - y^{(i)})$$.
So we can use the same general update rules from before
$$w \leftarrow w - \alpha (\hat{y}^i-y^i)x$$
$$b \leftarrow b - \alpha (\hat{y}^i-y^i)$$

Pictured below is the sigmoid function.

```python
# Define the sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Generate input values (x-axis)
x = np.linspace(-10, 10, 100)  # From -10 to 10 with 100 points

# Compute the sigmoid values (y-axis)
y = sigmoid(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y, label='Sigmoid Function', color='blue')
plt.title('Sigmoid Activation Function')
plt.xlabel('Input (x)')
plt.ylabel('Sigmoid Output')
plt.grid(True)
plt.legend()
plt.show()
```

# Reading in Data
Note that we are only considering player's games when they played for over a quarter to remove outlier performances.

```python
df = pd.read_csv("NBA-BoxScores-2023-2024.csv")
df = df.dropna(subset = ["START_POSITION"])
```

```python
# Function to convert time string to float
def time_to_float(time_str):
    mins, secs = time_str.split(':')
    return float(mins) + float(secs) / 60.0

# Apply the function to the 'MIN' column
df['MIN'] = df['MIN'].apply(time_to_float)

#Only consider players playing over a quarter
df = df[df["MIN"] >= 12]

#Creates a target column
df["isCenter"] = df["START_POSITION"] == "C"


#Normaizes stats for playing duration
for col in ["FG3A", "REB", "BLK","FGA"]:
    df[f"{col}_pm"] = df[col] / df["MIN"]

```

### Aggregating Player Season Stats
We are going to grab the average stats for a player across the season. This removes outlier performances and better reflects trends of players.

```python
ndf = df.groupby("PLAYER_ID").agg(
    FG3A_pm = ('FG3A_pm', 'mean'),
    REB_pm=('REB_pm', 'mean'),
    BLK_pm=('BLK_pm', 'mean'),
    FG_PCT = ("FG_PCT", "mean"),
    isCenter=("isCenter", "first"),
    FGA_pm=('FGA_pm', 'mean'),
)
```

We will reverify that our trends still hold in the truncated dataset.

```python
class SingleNeuron(object):
    def __init__(self, activation_function, cost_function):
        """
        Determines Activation and cost function -- for logistic regression
        this is sigmoid & cross entropy loss
        """
        self.activation_function = activation_function
        self.cost_function = cost_function

    def train(self, X, y, alpha = 0.005, epochs = 50):
        """
        Trains the model of features X and outputs y with a
        laerning rate alpha and a set number of epochs
        """
        self.w_ = np.random.randn(1 + X.shape[1])
        self.errors_ = []
        N = X.shape[0]

        for _ in range(epochs):
            errors = 0
            for xi, target in zip(X, y):
                error = (self.predict(xi) - target)
                self.w_[:-1] -= alpha*error*xi
                self.w_[-1] -= alpha*error
                #errors += .5*((self.predict(xi) - target)**2)
                errors += self.cost_function(self.predict(xi), target)
            self.errors_.append(errors/N)
        return self

    def predict(self, X):
        """
        Given some features, gives a probability that that data point
        is 1
        """
        preactivation = np.dot(X, self.w_[:-1]) + self.w_[-1]
        return self.activation_function(preactivation)

    def plot_cost_function(self):
        """
        Plots the cost over time
        """
        fig, axs = plt.subplots(figsize = (10, 8))
        axs.plot(range(1, len(self.errors_) + 1), 
                self.errors_,
                label = "Cost function")
        axs.set_xlabel("epochs", fontsize = 15)
        axs.set_ylabel("Cost", fontsize = 15)
        axs.legend(fontsize = 15)
        axs.set_title("Cost Calculated after Epoch During Training", fontsize = 18)
        plt.show()
        
def cross_entropy_loss(y_hat, y):
    return - y*np.log(y_hat) - (1 - y)*np.log(1 - y_hat)
```

```python
from sklearn.model_selection import train_test_split

X = ndf[["FG3A_pm", "REB_pm", "BLK_pm", "FG_PCT"]]  # Feature columns
y = ndf['isCenter']  # Target column

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=2)

model = SingleNeuron(sigmoid, cross_entropy_loss)
model.train(X_train.values, y_train, alpha = 0.01, epochs = 10_000)
```

```python
model.plot_cost_function()
```

### Evaluating the Model

```python
from sklearn.metrics import accuracy_score
predictions = np.round(model.predict(X_test))
actual = y_test.values

accuracy_score(actual, predictions)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(actual, predictions)
```
