# Center vs Guard
For our simplest algorithm (the perceptron), we will start with our simplest problem. Given an NBA player's height and weight, are they a center or guard. Centers are usually the tallest players on the team and point guards are usually the shortest. See NBA center Yao Ming (7-6) and guard Earl Boykins (5-5) below.  

The reason we are picking this problem is because the perceptron works best for clearly linearly seperable binary classifaction (ie you can draw a straight line between the two categories). Due to this limitation in the algorithm, the center / guard classifaction works well.

# The Perceptron
The Perceptron is a supervised learning algorithm used for binary classifcations. At its core, the algorithm takes in some features, applies some linear combination of these inputs with weights and a bias term, and then applies an activation faction to make a binary classifaction. After this, the algorithm checks its error and adjusts the weights and biases according to the Perceptron update rule and does this for a set number of iterations.
## Step 1
We take in $\vec{x}$, a vector of our input features. In our example, we have $x_1$ as a player's height and $x_2$ as a players weight. We will then generate our sample weights and bias according to a normal distribution $\vec{w},b\sim N(0,1)$. Then we will make our linear prediction
$$z = \vec{w}\cdot\vec{x}+b$$
and apply an activation function to $z$ in order to classify it. In our case, our two categories are $1$ and $-1$ so we will apply the following activation function to $z$,
$$
\phi (z)=\begin{cases} 
      1 & z \geq 0 \\
      0 & z\leq 0
   \end{cases}
$$

## Step 2
For each feature $\vec{x}^i$, we calculate $\hat{y}^i=\phi(z^i)$ and can evaluate performance with the  cost function
$$C(\vec{w},b)=\frac{1}{4}\sum_{i=1}^N(\hat{y}^i-y^i)$$
Then after each iteration, we update our weights and bias with the following rule (which is derived from minimizing the distance between the $y^i$ and $\hat{y}^i$ vectors.
$$w \leftarrow w - \eta(\hat{y}^i-y^i)x$$
and for the bias
$$b \leftarrow b - \eta(\hat{y}^i-y^i)$$
for some learning rate $\eta$.

# Cleaning up the data
To clean the data, we will do the following:
* Remove all rows with missing height or weight
* Only consider centers and guards
* Convert the height from feet & inches to just inches
* Only consider players who started playing after 2000
    * The reason for this, as we can see below, is that the average player height as increased a lot over time (ignoring minor variablity). In order to show off a clean linear classification, we will only focus on the modern era.

```python
df = df.dropna(subset=["height", "weight"])

def convert_to_inches(height):
    feet, inches = map(int, height.split('-'))
    return feet * 12 + inches

df['height_in_inches'] = df['height'].apply(convert_to_inches)

df.groupby("from_year").height_in_inches.mean().plot()
plt.title("Average Height Over Time")
plt.ylabel("Height (in)")
plt.xlabel("Debut Year")
plt.show()

df = df[(df["position"] == "Center") | (df["position"] == "Guard")]

df = df[df["from_year"] > 2000]
```

```python
df.boxplot(column='height_in_inches', by='position', grid=False)

plt.xlabel('Position')
plt.ylabel('Height (cm)')
plt.title('Height Distribution by Position')
plt.suptitle('')
plt.show()

df.boxplot(column='weight', by='position', grid=False)

plt.xlabel('Position')
plt.ylabel('Weight (lbs)')
plt.title('Weight Distribution by Position')
plt.suptitle('')
plt.show()
```

Here we can see the distributions of weight and height for both positions. While the average center is much taller and heavier than the average guard, the tallest guards are only slightly smaller than the smallest centers. These sharp distinctions should allow us to easily linearly seperate the data.

# Preprocessing the Data
This includes
* Scaling X values between 0 and 1
    * This helps the algorithm generalize better
* Converting positions to -1 and 1
* Creating a test train split


```python

X = df[["height_in_inches", "weight"]].values


y = df.position.values

#-1 for guards and 1 for centers
y = np.where(y == 'Guard', -1, 1)


X_scaled = (X - X.min(axis=0)) / (X.max(axis=0) - X.min(axis=0))

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, random_state = 8)

plt.scatter(X_train[:, 0], X_train[:,1], c=y_train)
plt.title("Height & Weight of Centers & Guards")
plt.xlabel('height')
plt.ylabel('weight')
plt.show()
```

Again, here we can see the gap between the centers (yellow) and the guards (purple). We will now implement the Perceptron class.

# Implementing the Perceptron Class

```python
class Perceptron:

    def __init__(self, eta=0.5, epochs=50):
        """Initialize the perceptron with learning rate and number of epochs."""
        self.eta = eta
        self.epochs = epochs

    def train(self, X, y):
        """
        Train the perceptron model using the given training dataset.
        X - features
        y - class
        """
        self.w_ = np.random.rand(1 + X.shape[1])  # Initialize weights randomly
        self.errors_ = []  # Track misclassifications

        for _ in range(self.epochs):
            errors = 0
            for xi, target in zip(X, y):
                update = self.eta * (self.predict(xi) - target)  # Compute weight update
                self.w_[:-1] -= update * xi  # Update feature weights
                self.w_[-1] -= update  # Update bias weight
                errors += int(update != 0)  # Track number of errors
                
            if errors == 0:
                return self  # Stop early if no errors
            else:
                self.errors_.append(errors)  # Store errors count
        
        return self

    def net_input(self, X):
        """
        Compute the net input (weighted sum of features plus bias).
        """
        
        return np.dot(X, self.w_[:-1]) + self.w_[-1]

    def predict(self, X):
        """
        Predict the class label based on net input.
        """
        return np.where(self.net_input(X) >= 0.0, 1, -1)
```

```python
# Instantiate one instance of the Perceptron class
clf = Perceptron(epochs = 50)

# Call the train method to train the weights and bias of the given instance
clf.train(X_train, y_train)


correct = 0
count = 0
for X_test_case, y_correct in zip(X_test, y_test):
    count += 1
    if clf.predict(X_test_case) == y_correct:
        correct += 1
print(f"Our classifier has a {correct/count} accuracy")
```

# Takeaways
Using our perceptron we trained on the train set, we can come up with the following linear categorization of the data.

```python
plt.figure(figsize = (10, 8))
plot_decision_regions(X_scaled, y, clf = clf)
plt.title("Perceptron Outputs", fontsize = 18)
plt.xlabel("Player height [in]", fontsize = 15)
plt.ylabel("Player weight [lbs]", fontsize = 15)
plt.show()
```

The perceptron is perfectly able to distinguish centers from guards. While this is a simply example meant to demonstrate linearly classifying data, the true power of the perceptron comes when we modify it for linear regression, logistic regression.
